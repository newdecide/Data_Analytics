```python
# 14:22~16:25
from bs4 import BeautifulSoup as bs
```

```python
with open("./VC.html",'r') as f:
    fileread = f.read()
print(fileread)
```

```python
import re
result = bs(fileread, 'html.parser')
str1 = str(result.find_all("div"))

# 태그 지우는 코드(re 정규표현식)
OMG = re.sub('<div.*?>.*?</div>', '', str1, 0, re.I|re.S)
result2 = bs(OMG, 'html.parser')

print(result2)
```

```python
result = []
for el in result2.find_all("ul",{'class':'section-list'}):
#   print(el.text.split())
    for x in el.find("a",{'class': "item"}).text.split():
        st = ''.join(list(x))
        result.append(st)
        #     #     for el2 in el.find("a", {'class': 'item'}):
#         print(el2)
print(result)
```

```python
for el in result.find_all("ul",{'class':'section-list'}):
#     print(el.text.strip('Start'))
    
    content = el.text
    print(type(content))
    print(type(el))
```
