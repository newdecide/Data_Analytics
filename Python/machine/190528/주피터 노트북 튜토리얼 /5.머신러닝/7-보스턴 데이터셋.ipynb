```python
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

boston = load_boston()
df1 = pd.DataFrame(boston['data'],
                  columns = boston['feature_names'])
df1['target'] = boston['target']
df1

x = df1.values[:,:13] # 12까지
y = df1.values[:,13] # 13

#random_state 를 주어서 고정을 함.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

```

```python
result2 = dict()
model = dict()

from sklearn  import linear_model

reg = linear_model.Ridge() # 선형 많이 쓰는 함수
reg.fit(x_train, y_train)
reg.score(x_train, y_train)
result2['Ridge'] = reg.score(x_test, y_test)
model['Ridge'] = reg


reg = linear_model.Lasso() # 
reg.fit(x_train, y_train)
reg.score(x_train, y_train)
result2['Lasso'] = reg.score(x_test, y_test)
model['Lasso'] = reg


from sklearn.svm import SVR
reg = SVR()
reg.fit(x_train, y_train)
result2['SVR'] = reg.score(x_test, y_test)
model['SVR'] = reg

from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(x_train, y_train)
result2['Tree'] = reg.score(x_test, y_test)
model['Tree'] = reg

import pandas as pd
pd.Series(result2)
```

```python
pred_y = model['Ridge'].predict(x_test)
pred_y
```
