```python
import pandas as pd
```

```python
data = {
    'id' : [1,2,3],
    'name' : ['choi', 'Lee', 'Kim'],
    'age' : [10,20,30],
    'assets': [150.4, 123.4, 88.88],
    'job': ['student', 'CEO', 'Dad']
}

df = pd.DataFrame(data)
df
```

```python
#연습문제

#구구단 1단부터 3단까지 출력하는 데이터 프레임 만들기
dan1 = pd.Series(range(10))
dan2 = dan1 *2
dan3 = dan1 *3

data = {
    '1단' : dan1,
    '2단' : dan2,
    '3단' : dan3
}

df1 = pd.DataFrame(data)
df1
```

```python
#인덱스명 바꾸기
df1.index = ['영','일', '이', '삼', '사', '오','육', '칠','팔','구']
df1
```

```python
df1['4단'] = df1['1단'] * 4
df1
```

```python
df1['1단'] = df1['1단'] *11
```

```python
#컬럼명 변경 
df1.columns = ['11단', '2단', '3단', '4단']
df1
```

```python
#11단 삭제
df1.drop('11단', axis = 1)
#역행렬
df1.T
```

```python
df1.drop(['오','칠'])
df1
```
