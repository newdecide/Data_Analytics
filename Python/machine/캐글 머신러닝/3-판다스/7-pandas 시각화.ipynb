```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#100개의 랜덤한 값을 넣는 것
se = pd.Series(np.random.randn(100).cumsum())

#메모리 상에서만 보여줌. 플롯 안에 넣어야 바뀜. 카인드로 원 안에 그림을 넣을 수 있다. 
se.plot()
```

```python
df = pd.DataFrame(np.random.randn(100, 5).cumsum(0), 
                     columns = ['a1', 'a2', 'a3', 'a4', 'a5'] 
                 )
df
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'barh')
```

```python
#데이터의 합을 보는 것 / 남녀의 합을 구할 때 주로 사용함.
df.plot(kind = 'bar', stacked = True)
```

```python
# 붓꽃 데이터 
# 어떤 꽃인지 확인하는 것.
from sklearn import datasets
iris =datasets.load_iris()
irisdt = pd.DataFrame(iris['data'], columns = iris['feature_names'])
irisdt
```

```python
plt.scatter(irisdt['sepal length (cm)'], irisdt['petal length (cm)'],
           c= iris['target'])
```

```python
pd.plotting.scatter_matrix(irisdt, diagonal = 'kde', alpha= 0.3)
```

# 머신러닝 맛보기
모델이 구현이 됨. => 적합한 모델로 성장 시킬 수 있을지 고민(튜닝)
KNN

```python
from sklearn import neighbors

#회귀
#neighbors.KNeighborsRegressor

#분류
knn = neighbors.KNeighborsClassifier()

#훈련작업 (기출문제와 전 문제가 달라야함)
knn.fit(iris['data'], iris['target'])
knn.score(iris['data'], iris['target'])
```
