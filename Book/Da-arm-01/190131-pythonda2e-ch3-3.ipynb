```python
#Day6 15:45 ~16:00
import pandas as pd
import numpy as np
import quandl
```

```python
sunspots = quandl.get("SIDC/SUNSPOTS_A")
```

```python
sunspots
```

```python
#head 맨위 부터, tail은 맨 아래 부터
print("Head 2", sunspots.head(2))
print("####")
print("Tail 2", sunspots.tail(2))
```

```python
#마지막 날짜를 사용해 마지막 값을 검색
last_date = sunspots.index[-1]
print("Last value\n", sunspots.loc[last_date])
```

```python
#날짜형식으로 검색
print("Values slice by date:\n", sunspots["20020101": "20131231"])
```

```python
#16:45 ~ 17:25
print("Slice from a list of indices:\n", sunspots.iloc[[2,4,-4,-2]])
```

```python
#통계표
print("Describe", sunspots.describe(),"\n")
```

```python
#카운트
print("count", sunspots.count(), "\n")
```

```python
#표준 편차
print(sunspots.mad(), "\n")
```

```python
#중앙값
print(sunspots.median(), "\n")
#최댓값
print(sunspots.max(), "\n")
#최솟값
print(sunspots.min(), "\n")
#최빈값
print(sunspots.mode(), "\n")
```

```python
#분산의 제곱근
print(sunspots.std(), "\n")
#분산을 반환
print(sunspots.var(), "\n")
```

```python
#비대칭 반환 
print(sunspots.skew())
```

```python
#첨도
print(sunspots.kurt())
```
