```python
#로그 플롯(logarithmic plot)
#로그 눈금(scale)

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('ex/transcount.csv')
df = df.groupby('year').aggregate(np.mean)
years = df.index.values
counts = df['trans_count'].values

#데이터 피팅
poly = np.polyfit(years, np.log(counts), deg=1)
#다항식 형태 결과
print("Poly", poly)

#semilogy 함수로 피팅 
plt.semilogy(years, counts, 'o')
plt.semilogy(years, np.exp(np.polyval(poly, years)))
plt.show()
```

```python
#분산 플롯(scatter plot) => 변수 사이의 상관관계를 보여준다. 
#양 positive
#bubble chart는 분산 플롯의 확장된 형태이다. 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('ex/transcount.csv')
df

df = df.groupby('year').aggregate(np.mean)

gpu = pd.read_csv('ex/gpu_transcount.csv')
gpu = gpu.groupby('year').aggregate(np.mean)

df = pd.merge(df, gpu, how='outer', left_index=True, right_index =True)

df = df.replace(np.nan, 0)
print(df)

years = df.index.values
counts = df['trans_count'].values
gpu_count = df['gpu_trans_count'].values

cnt_log = np.log(counts)

#알파는 투명도를 뜻함 
plt.scatter(years, cnt_log, c =200 * years, s = 20 +200 * gpu_count /gpu_count.max(), alpha = 0.5)
plt.show()
```
